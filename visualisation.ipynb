# Pandas Visualisation

## In This Practical, We are Using a Simple Ice Cream Dataset To compare Ice Cream Quality
#### Lets Dive in

```python
#Importing Of Necesarry Libraries to do visualization.
#Pandas mainly deal with the dataset.
import pandas as pd 
import numpy as np
#Matplot lib deals with the plotting, It is the plotting library.
import matplotlib.pyplot as plt
%matplotlib inline
```

### Reading A CSV File

```python
df = pd.read_csv('Ice Cream Ratings.csv')
# This will make our index to be date not the figures.
df = df.set_index('Date')
df
```

```python
# We are ploting all the features in one plotting area
df.plot.line()
plt.title('Ice Cream Ratings') #Giving our plot a tittle
plt.xlabel('Daily Ratings') # Giving Our plot an X label
plt.ylabel('Ratings') #Labeling our y axis
plt.show() #Displaying our plot
```

```python
# to further plot these columns by item.
df.plot(kind = 'line',subplots = True)
plt.title('Ice Cream Ratings')
plt.xlabel('Daily Ratings')
plt.ylabel('Ratings')
plt.show(1)
```

### Ploting the Bar Graph

```python
df.plot.bar(stacked = True)
plt.title('Ice Cream Ratings')
plt.xlabel('Daily Ratings')
plt.ylabel('Ratings')
plt.show()
```

```python
# KNow for individual ratings  eg Flavour Rating
df['Flavor Rating'].plot(kind='bar')
plt.show()
#same applies o all the others.
```

#### Horizontal Bar Chat

```python
# Here we use the barh command.
#The stacked command makes alll of the rating to appear on one tab
df.plot.barh(stacked = True)
plt.show()
```

### Ploting the Scatter Plot

```python
# Scatter Plots
df.plot.scatter(x = 'Texture Rating',y = 'Overall Rating',color ='green',marker='o')
plt.show(1)
```

### Ploting the Histogram

```python
df.plot.hist(bins= 20)
#Bins deal with the distribution of those values on the histogram
plt.show()
```

### Ploting The Box Plot

```python
df.boxplot()
plt.show()
```

### This is An Area Plot

```python
df.plot.area(figsize =(10,4))
# The attribute fig size deals with how big you want your image to be
#The first figure is for width, the second for height
plt.show()
```

### Ploting the Pie Chart

```python
df.plot.pie(y = 'Flavor Rating',figsize=(6,6))
plt.legend(loc = 'best')
plt.show()
```
